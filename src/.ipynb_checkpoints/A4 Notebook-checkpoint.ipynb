```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Getting data about case counts, mask mandates and compliance from 3 CSV files
confirmed_cases = pd.read_csv('../data/CovidRawData/covidcases/RAW_us_confirmed_cases.csv')
mask_mandates = pd.read_csv('../data/CovidRawData/maskdata/U.S._State_and_Territorial_Public_Mask_Mandates_From_April_10__2020_through_August_15__2021_by_County_by_Day.csv')
masks_compliance = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/mask-use/mask-use-by-county.csv')
```

```python
# Calculating a new FIPS column which is a combined representation of State and County. 
# Adding trailing 0s where necessary for single digit State codes.

modified_state_fips = mask_mandates['FIPS_State'].apply(lambda state_code: state_code * 100 if state_code < 9 else state_code * 1000)
mask_mandates['FIPS'] = mask_mandates['FIPS_County'] + modified_state_fips
# Changing date column to datetime
mask_mandates['date'] = pd.to_datetime(mask_mandates['date'], format = '%m/%d/%Y')

# Pivoting the confirmed_cases dataset to long format
cases = pd.melt(confirmed_cases, id_vars = confirmed_cases.columns[0:11], var_name = 'date', value_name = 'cases')
cases['date'] = pd.to_datetime(cases['date'], format = '%m/%d/%y')
```

```python
# Creating merged dataframe
merged_df = cases.merge(mask_mandates, on=['date','FIPS'], how = 'left')
merged_df = merged_df.merge(masks_compliance, right_on='COUNTYFP', left_on='FIPS', how = 'left')
covid_combined = merged_df.drop(['COUNTYFP', 'code3', 'Country_Region', 'iso2', 'iso3'], axis = 1)
```

```python
# Get Clark County, Nevada FIPS code
clarkcounty_nevada_fips = confirmed_cases.loc[(confirmed_cases['Admin2'] == 'Clark') & (confirmed_cases['Province_State'] == 'Nevada'), 'FIPS'].values[0]

# Filter Clark County data to correct time period between 2020/02/01 and 2021/10/15
filtered_clark_county = covid_combined.loc[(covid_combined['FIPS'] == clarkcounty_nevada_fips) &
                                           (covid_combined['date'] <= '2020-09-01') & 
                                           (covid_combined['date'] >= '2020-05-01')
                                          ]
clark_county = filtered_clark_county.rename({'Face_Masks_Required_in_Public': 'public_masks'}, axis = 1).sort_values(by = ['date'])

# Make public_masks boolean instead of 'Yes' and np.nan
clark_county['public_masks'] = clark_county['public_masks'] == 'Yes'
clark_county
```

We now want to check if we can calculate the potential Covid19 case count in Clark County if mask mandates were not enforced. For the purposes of this calculation, We will try to view mask mandates as unnecessary hassles that only reduce the spread of covid a 55%. Then the Covid cases without masks = masked cases/ 0.55. We also assume that the "contagious window" of time when a person is infected with Covid19 is 10 days and that the population that is at risk of infection is simply current population - population infected with Covid19 (no re-infections).

```python
mask_efficacy = 0.55
transmission_period = 10
clark_county_population = 3186989 # from 2020 Census data about Clark County.

clark_county['daily_cases_masked'] = clark_county['cases'].diff(periods = 1)
clark_county['daily_cases_unmasked'] = np.where(clark_county['public_masks'],
                                                clark_county['daily_cases_masked'] / mask_efficacy,
                                                clark_county['daily_cases_masked'])

clark_county['infectious_cases'] = clark_county['daily_cases_masked'].rolling(window = transmission_period).sum().fillna(0)
clark_county['infectious_cases_unmasked'] = clark_county['daily_cases_unmasked'].rolling(window = transmission_period).sum().fillna(0)

# Population-at-risk = Population - infectious cases
clark_county['popultion_at_risk'] = clark_county_population - clark_county['infectious_cases']
clark_county['population_at_risk_unmasked'] = clark_county_population - clark_county['infectious_cases_unmasked']

# Infection rate = (number of daily infections) / (daily population at risk)
clark_county['infection_rate'] = (clark_county['daily_cases_masked'] / clark_county['population_at_risk']) * 100.0
clark_county['infection_rate_unmasked'] = (clark_county['daily_cases_unmasked'] / clark_county['population_at_risk_unmasked']) * 100.0
```

```python
fig = plt.figure(figsize = (23, 10))
plt.plot(clark_county['date'], clark_county['daily_cases_masked'],
        label = 'Confirmed daily cases',
        color = 'red')
plt.plot(clark_county.loc[clark_county['public_masks'], 'date'], clark_county.loc[clark_county['public_masks'], 'daily_cases_masked'],
        label = 'Confirmed daily cases with mask mandate enforced',
        color = 'darkgreen')
plt.plot(clark_county.loc[clark_county['public_masks'], 'date'], clark_county.loc[clark_county['public_masks'], 'daily_cases_unmasked'],
         label = 'Estimated daily cases without mask mandate enforced',
         color = 'purple', linestyle = '--')

plt.legend(fontsize = 12)
plt.grid()
plt.xlabel('Date', fontsize = 12)
plt.ylabel('Number of daily cases', fontsize = 12)
plt.title('Progression of confirmed COVID-19 cases in Clark County, Nevada\n From February 1, 2020 through October 15, 2021', fontsize = 15)
```

```python
fig = plt.figure(figsize = (23, 10))
plt.plot(clark_county['date'], clark_county['infection_rate'],
         label = 'Confirmed daily infection rate',
         color = 'red', linewidth = 2)
plt.plot(clark_county.loc[clark_county['public_masks'], 'date'], clark_county.loc[clark_county['public_masks'], 'infection_rate'],
         label = 'Confirmed daily infection rate with mask mandate enforced',
         color = 'darkgreen', linewidth = 2)
plt.plot(clark_county.loc[clark_county['public_masks'], 'date'], clark_county.loc[clark_county['public_masks'], 'infection_rate_unmasked'],
         label = 'Estimated daily infection rate without mask mandate enforced',
         color = 'purple', linestyle = '--')

plt.legend(fontsize = 12)
plt.grid()
plt.xlabel('Date', fontsize = 12)
plt.ylabel('Confirmed daily COVID-19 infection Rate %', fontsize = 12)
plt.title('Confirmed daily COVID-19 infection rate % in Clark County, Nevada\n From February 1, 2020 through October 15, 2021', fontsize = 15)
```
